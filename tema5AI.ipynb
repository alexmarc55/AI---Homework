```python
import json
import random
import numpy as np


with open('problem_2.json', 'r') as f:
    data = json.load(f)
```

```python
projects = data['projects']
budget_total = data['budget_total']
time_total = data['time_total']
values = [project['value'] for project in projects]
costs = [project['cost'] for project in projects]
times = [project['time'] for project in projects]
```

```python
def fitness(chromosome, values, costs, times, max_budget, max_time):
    total_value = sum(v * x for v, x in zip(values, chromosome))
    total_cost = sum(c * x for c, x in zip(costs, chromosome))
    total_time = sum(t * x for t, x in zip(times, chromosome))

    # penalizare daca depasim constrangerile
    if total_cost > budget_total or total_time > time_total:
        penalty = 0.1
        return total_value * penalty
    return total_value
```

```python
def generate_population(n_projects, pop_size):
    return [np.random.randint(0, 2, n_projects).tolist() for _ in range(pop_size)]

def tournament_selection(pop, fitnesses, k=3):
    candidates = random.sample(range(len(pop)), k)
    return max(candidates, key=lambda i: fitnesses[i])

def single_point_crossover(p1, p2):
    point = random.randint(1, len(p1) - 1)
    return p1[:point] + p2[point:], p2[:point] + p1[point:]

def mutate(chromosome, rate=0.1):
    for i in range(len(chromosome)):
        if random.random() < rate:
            chromosome[i] = 1 - chromosome[i]
    return chromosome
```

```python
def run_ga(values, costs, times, max_budget, max_time,
           pop_size=20, generations=100, mutation_rate=0.1):

    n_projects = len(values)
    population = generate_population(n_projects, pop_size)

    for gen in range(generations):
        fitnesses = [fitness(ind, values, costs, times, max_budget, max_time) for ind in population]
        new_pop = []

        # elitism: pastram cel mai bun individ
        best_idx = np.argmax(fitnesses)
        new_pop.append(population[best_idx])

        while len(new_pop) < pop_size:
            p1 = population[tournament_selection(population, fitnesses)]
            p2 = population[tournament_selection(population, fitnesses)]

            c1, c2 = single_point_crossover(p1, p2)
            c1 = mutate(c1, mutation_rate)
            c2 = mutate(c2, mutation_rate)

            new_pop.extend([c1, c2])

        population = new_pop[:pop_size]

    # rezultat final
    fitnesses = [fitness(ind, values, costs, times, budget_total, time_total) for ind in population]
    best_idx = np.argmax(fitnesses)
    best = population[best_idx]
    total_value = sum(v * x for v, x in zip(values, best))
    total_cost = sum(c * x for c, x in zip(costs, best))
    total_time = sum(t * x for t, x in zip(times, best))

    print("\n=== REZULTAT FINAL ===")
    print(f"Cel mai bun cromozom: {best}")
    print(f"Valoare totala: {total_value}")
    print(f"Cost total: {total_cost}/{budget_total}")
    print(f"Timp total: {total_time}/{time_total}")
    return best, total_value, total_cost, total_time
```

```python
best, total_value, total_cost, total_time = run_ga(values, costs, times, budget_total, time_total, pop_size=20, generations=2000, mutation_rate=0.1)
```
